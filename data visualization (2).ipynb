1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Output to notebook
output_notebook()

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Show the plot
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

# Output to notebook
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10
sizes = np.random.rand(n) * 20  # Marker sizes
colors = np.random.rand(n)  # Colors

# Create a DataFrame (optional, but useful for handling column data)
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a new plot
p = figure(title="Scatter Plot with Varying Sizes and Colors", x_axis_label='x', y_axis_label='y', width=800, height=400)

# Add a scatter renderer
p.scatter(x='x', y='y', size='sizes', color='colors', source=data, legend_label="Random Data", fill_alpha=0.6, line_color=None)

# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

   fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
   counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import pandas as pd

# Output to notebook
output_notebook()

# Define the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a DataFrame
data = pd.DataFrame({'fruits': fruits, 'counts': counts})

# Create a new plot
p = figure(x_range=data['fruits'], title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', width=800, height=400)

# Add a bar renderer
p.vbar(x='fruits', top='counts', width=0.4, source=data, legend_label="Counts", color="orange", alpha=0.7)

# Customize the plot
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.xaxis.major_label_orientation = "vertical"

# Show the plot
show(p)

```

4. Create a Bokeh histogram to visualize the distribution of the given data

   data_hist = np.random.randn(1000)
   hist, edges = np.histogram(data_hist, bins=30)

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Output to notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a new plot
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency', width=800, height=400)

# Add a quad renderer
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Show the plot
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.


data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import numpy as np


# Output to notebook
output_notebook()

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the arrays for Bokeh
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Create a DataFrame
data = pd.DataFrame({'x': x_flat, 'y': y_flat, 'value': z_flat})

# Create a new plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=800, height=400, tooltips=[("X", "@x"), ("Y", "@y"), ("Value", "@value")])

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(z_flat), high=max(z_flat))

# Add a rect renderer
p.rect(x='x', y='y', width=0.1, height=0.1, source=data, fill_color=mapper, line_color=None)

# Show the plot
show(p)
```
