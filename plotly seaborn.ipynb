```python
#1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.


import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate synthetic data
np.random.seed(0)  # For reproducibility
n = 100
x = np.random.rand(n) * 100  # Random x values
y = 2.5 * x + np.random.randn(n) * 10  # y values with a linear relationship and some noise

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y})

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='x', y='y', data=data, color='blue', alpha=0.7)

# Customize the plot
plt.title('Scatter Plot of Synthetic Data')
plt.xlabel('X')
plt.ylabel('Y')

# Show the plot
plt.show()

```

```python
#2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.


# Generate random data
np.random.seed(42)  # For reproducibility
data = np.random.randn(1000)  # 1000 random numbers from a normal distribution

# Create a DataFrame
import pandas as pd
df = pd.DataFrame({'value': data})

# Create a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['value'], bins=30, kde=True, color='skyblue')

# Customize the plot
plt.title('Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()

```

```python
#3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.



# Define the dataset
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [150, 200, 180, 220]

# Create a DataFrame
data = pd.DataFrame({'Category': categories, 'Value': values})

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value', data=data, palette='viridis')

# Customize the plot
plt.title('Comparison of Categories Based on Values')
plt.xlabel('Category')
plt.ylabel('Value')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```

```python
#4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.


import matplotlib.pyplot as plt


# Generate synthetic data
np.random.seed(42)
categories = ['Category A', 'Category B', 'Category C', 'Category D']
data = {
    'Category': np.repeat(categories, 100),
    'Value': np.concatenate([np.random.normal(loc=mean, scale=10, size=100) for mean in [50, 60, 70, 80]])
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='viridis')

# Customize the plot
plt.title('Distribution of Values Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')

# Show the plot
plt.show()
```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.


# Generate synthetic data
np.random.seed(42)
n_samples = 100
n_features = 4

# Create correlated features
# Base feature
x = np.random.rand(n_samples)
# Other features are correlated with x
data = {
    'Feature 1': x,
    'Feature 2': x + np.random.normal(0, 0.1, n_samples),
    'Feature 3': x * 2 + np.random.normal(0, 0.1, n_samples),
    'Feature 4': np.random.rand(n_samples) * 0.5 + np.random.normal(0, 0.1, n_samples)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Compute the correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, linewidths=0.5, fmt=".2f")

# Customize the plot
plt.title('Correlation Matrix Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=45)

# Show the plot
plt.show()
```

```python
import plotly.graph_objs as go
import pandas as pd
import numpy as np

# Generate the dataset
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=df['X'],
    y=df['Y'],
    z=df['Z'],
    mode='markers',
    marker=dict(
        size=5,
        color='blue',
        opacity=0.6
    )
)])

# Customize the layout
fig.update_layout(
    title='3D Scatter Plot of Random Data',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Generate the dataset for sales
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a matrix suitable for a heatmap
sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create a heatmap
fig = go.Figure(data=go.Heatmap(
    z=sales_pivot.values,
    x=sales_pivot.columns,
    y=sales_pivot.index,
    colorscale='Viridis',
    colorbar=dict(title='Sales')
))

# Customize the layout
fig.update_layout(
    title='Sales Variation Across Months and Days',
    xaxis_title='Month',
    yaxis_title='Day'
)

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Generate the data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the arrays and create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(
    z=z,
    x=x,
    y=y,
    colorscale='Viridis'
)])

# Customize the layout
fig.update_layout(
    title='3D Surface Plot of Sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Generate the dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
fig = go.Figure(data=[go.Scatter(
    x=df['GDP'],
    y=df['Population'],
    mode='markers',
    text=df['Country'],  # Display country names when hovering over bubbles
    marker=dict(
        size=df['Population'] / 10,  # Scale the bubble size (adjust divisor as needed)
        color=df['Population'],  # Color bubbles based on population
        colorscale='Viridis',
        showscale=True
    )
)])

# Customize the layout
fig.update_layout(
    title='Bubble Chart of Countries by Population and GDP',
    xaxis_title='GDP',
    yaxis_title='Population',
    xaxis=dict(type='log'),  # Use log scale if GDP varies widely
    yaxis=dict(type='log')   # Use log scale if Population varies widely
)

# Show the plot
fig.show()
```
